`The Creation of Execution Context`

When you run a JavaScript program, a lot of things happen behind the scenes inside the JavaScript engine. One of the key concepts in JavaScript is the execution context, which is responsible for everything that happens in the program.

The execution context is created in two phases: the memory creation phase and the code execution phase.



`Memory Creation Phase`

In the memory creation phase, JavaScript allocates memory to all the variables and functions in the program.

For variables, JavaScript reserves memory space and assigns a special value called "undefined".

For functions, JavaScript stores the entire code of the function in the allocated memory space.


`Code Execution Phase`

In the code execution phase, JavaScript executes the code line by line.

When a variable is encountered, JavaScript replaces the "undefined" value with the actual value.

When a function is invoked, JavaScript creates a new execution context and executes the code within that function.

`Call Stack: Managing Execution Contexts`

JavaScript uses a data structure called the call stack to manage the execution contexts.

The call stack maintains the order of execution of execution contexts. When a new execution context is created, it is pushed onto the call stack. When an execution context is finished executing, it is popped off the call stack.

The call stack is also known as the execution context stack, program stack, control stack, runtime stack, or machine stack.

Conclusion

Understanding the behind-the-scenes workings of the execution context is crucial for mastering JavaScript. The execution context is responsible for managing variables, functions, and the flow of code execution. By understanding how the execution context is created and managed, you can write more efficient and effective JavaScript code.

Next thing is Hoisting  in JS ?

    The javascript mechanism in which variables and function declarations are moved to the top of their scope before execution of the code is called Hoisting.

Takeaway:

    Hoisting is a mechanism that makes it seem like variables and function declarations are moved to the top of the code and lets us access variables and functions before they are declared.

Creation of execution context

How Javascript executes the code behind the scenes is that it creates a (global) execution context. This execution context creation has two phases: Creation and Execution

1) Creation phase:

Javascript engine first scans the code, registers the variables in the scope, and allocates memory to all the variable and function declarations.
If it encounters a variable, it is given the value undefined, and if it encounters a function, then the value is the function code itself, and a new (local) execution context is created.


2) Execution phase:

After the creation phase, the JS engine executes the code line by line.
So in our previous example:



```python
console.log(temp)
var temp = "Scaler"
```


the temp variable will first be assigned the value undefined, and then the execution happens line by line.

Takeaway:

Javascript Engine executes code in two phases: The creation phase and the execution phase. In the creation phase, the variables are registered in the scope; this is what makes Hoisting possible, while the execution phase comes after the creation phase, where the JS engine executes the code line by line.

Undefined vs ReferenceError
In the examples under the intro, we have seen both of these terms. Let's clearly understand them here.

Undefined:

As mentioned previously, undefined is the value allocated to a variable during the creation phase. It is one of the primitive data types in javascript. A variable can have the value undefined when it is declared as being this type i.e. when we write var a = undefined or when the value undefined is assigned to it, similar to what we have seen in the creation phase.

ReferenceError:

Reference error occurs when we try to access a variable that is not declared. This also explains why in the previous example,

console.log(temp)

we got the reference error because this time, there was no variable declaration at all! Hence, the appropriate error was given.

Takeaway:

A variable has the value undefined when it is declared with this data type or assigned with it. Reference error arises when we try to access a variable before declaring it.

let and const Hoisting in JavaScript

let and const was introduced in ES6. Is hoisting possible with let and const? Though in many places you will see that the answer is No. This is not true All javascript variables are hoisted irrespective of var or let or const declaration, but let and const are hoisted differently. Let us look in-depth into this in the next sections.

Takeaway:

let and const are hoisted in Javascript but they are hoisted differently than var



```python
console.log(temp)
let temp= "Scaler"
```

for the above code  it give you error 

Uncaught ReferenceError: Cannot access 'temp' before initialization at ++script.js.1++


We got an error that we did not see in the previous examples.

The reason for this is var is both declared and initialized during Hoisting, but let is not initialized it is only declared. The variable is in a temporal dead zone(TDZ) till then.

Let us see the let variable lifecycle to understand this better:




<img alt="let variable lifecycle" loading="lazy" width="864" height="200" decoding="async" data-nimg="1" class="markdown_image_container__jjYk1 markdown_loaded__iZ4O5" src="https://scaler.com/topics/images/let-variable-lifecycle.webp" style="color: transparent; height: auto; max-width: 100%;">






Explanation:

In the creation phase for variables declared with let only declaration occurs there is no initialization.
A variable declared with let will be initialized only during the execution phase when the JS engine reaches the line with let keyword at this point it will be given the value undefined
The gap we see between the point at which the variable is in the uninitialized state and the point where it gets initialized is called the Temporal dead zone (TDZ)
Then the assignment happens temp is given the value "Scaler".






`Takeaway`:

var is both declared and initialized during Hoisting, but let is not initialized it is only declared. Hence when we try to access a let variable before it is declared we get an error.



`const Keyword Hoisting in JavaScript`


const hoisting is the same as let hoisting except for one difference. In let the assignment happens after initialization but in const the assignment and initialization happen together, this is because once assigned the value of const cannot change if we assign undefined and then assign another value this breaks the rule of const



```python
console.log(temp)
const temp = "Scaler"
```

for the above code it show  this error

`Conclusion`


The javascript mechanism in which variables and function declarations are moved to the top of their scope before code execution is called Hoisting in Javascript


Hoisting gives us an advantage in that variables and functions can be accessed before they are declared.


Function expressions and arrow functions cannot be hoisted.


The sequence of variable declaration and initialization or the lifecycle of a variable is as follows: Declaration -> Initialization -> Assignment.
